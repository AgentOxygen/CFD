```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.colors as colors
from matplotlib import rc
from scipy.optimize import fsolve
from scipy.integrate import solve_ivp
from numpy.polynomial import polynomial as poly
from scipy.optimize import curve_fit

def euler_explicit(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0

    for i in range(0, len(x) - 1):
        y[i+1] = y[i] + h*f(x[i], y[i])
    return y


def euler_implicit(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    
    for i in range(1, len(x) - 1):
        residual = lambda yi : yi - y[i] - h*f(x[i+1], yi)
        y_solved = fsolve(residual, y[i] + h*f(x[i], y[i]))
        y[i+1] = y[i] + h*f(x[i+1], y_solved)
    return y


def midpoint(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0

    for i in range(0, len(x) - 1):
        y1 = f(x[i], y[i])
        y2 = f(x[i] + h/2, y[i] + (h/2)*y1)
        y[i+1] = y[i] + h*y2
    return y


def trapezoidal(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    
    for i in range(0, len(x) - 1):
        residual = lambda yi : yi - y[i-1] - h*f(x[i+1], yi)
        y_solved = fsolve(residual, y[i] + h*f(x[i], y[i]))
        
        y[i+1] = y[i] + (h/2)*(f(x[i], y[i]) + f(x[i+1], y_solved))
    return y

def adams_bashforth2(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    y[1] = y[0] + h*f(x[0], y[0])
    
    for i in range(1, len(x) - 1):
        y[i+1] = y[i] + h*((3/2)*f(x[i], y[i]) - (1/2)*f(x[i-1], y[i-1]))
    return y

# def explicit_RK2(x, f, h, y0):
#     y = np.zeros(x.shape)
#     y[0] = y0
    
#     for i in range(0, len(x) - 1):
#         xi = x[i] + (h/2)
#         yi = y[i] + (h/2)*f(x[i], y[i])
#         y[i+1] = y[i] + h*f(xi, yi)
#     return y


def explicit_RK4(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    
    for i in range(0, len(x) - 1):
        y1 = f(x[i], y[i])
        y2 = f(x[i] + (h/2), y[i] + (h/2)*y1)
        y3 = f(x[i] + (h/2), y[i] + (h/2)*y2)
        y4 = f(x[i] + h, y[i] + h*y3)
        y[i+1] = y[i] + (h/6)*(y1 + 2*y2 + 2*y3 + y4)
    return y

####### Solver Parameters #######
dydx = lambda x, y : -50*(y - np.cos(x))
h_step = 10**(-2)
x_pts = np.arange(0, 1 + h_step, h_step)
y_init = 0
```

```python
####### Execute Solvers #######
actual = solve_ivp(dydx, [0,1], [y_init], t_eval=x_pts)
e_exp = euler_explicit(x_pts, dydx, h_step, y_init)
e_imp = euler_implicit(x_pts, dydx, h_step, y_init)
mdpt_method = midpoint(x_pts, dydx, h_step, y_init)
trap = trapezoidal(x_pts, dydx, h_step, y_init)
adam = adams_bashforth2(x_pts, dydx, h_step, y_init)
rk2 = explicit_RK2(x_pts, dydx, h_step, y_init)
rk4 = explicit_RK4(x_pts, dydx, h_step, y_init)

####### Plot Results #######
f, ax1 = plt.subplots(1, 1, figsize=(14, 8), facecolor='w')
f.suptitle(f"Various ODE Solvers with Step 'h = {h_step}'", fontsize=30)
ax1.tick_params(axis='both', which='major', labelsize=15)

lw=5
ls="-"
alpha=0.75
ax1.plot(x_pts, e_exp, color="Blue", label="Explicit Euler", linestyle=ls, linewidth=lw, alpha=alpha)
ax1.plot(x_pts, mdpt_method, color="Green", label="Midpoint", linestyle=ls, linewidth=lw, alpha=alpha)
ax1.plot(x_pts, e_imp, color="Red", label="Implicit Euler", linestyle=ls, linewidth=lw, alpha=alpha)
lw=2
ls="--"
alpha=1
ax1.plot(x_pts, trap, color="Orange", label="Trapezoidal", linestyle=ls, linewidth=lw, alpha=alpha)
ax1.plot(x_pts, adam, color="Purple", label="Adams-Bashforth 2", linestyle=ls, linewidth=lw, alpha=alpha)
ax1.plot(x_pts, rk2, color="Cyan", label="Explicit RK2", linestyle=ls, linewidth=lw, alpha=alpha)
ax1.plot(x_pts, rk2, color="Lime", label="Explicit RK4", linestyle=ls, linewidth=lw, alpha=alpha)
ax1.plot(actual.t, actual.y[0], color="Black", label="Actual (scipy.solve_ivp)", linestyle=ls, linewidth=lw, alpha=alpha)

ax1.grid()
ax1.legend(fontsize=15)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_xlabel("x", fontsize=20)
ax1.set_ylabel("y", fontsize=20)
```

```python
f, axes = plt.subplots(4, 2, figsize=(16, 20), facecolor='w')
f.suptitle(f"Global Error for Solvers at Different Step Sizes", fontsize=30, y=1.0)

#h_steps = np.array([2, 5, 20, ])
h_steps = np.array([100, 200, 500, 1000, 2000, 5000])
#h_steps = np.array([i for i in range(100, 1000, 10)])

methods = [
    (solve_ivp, "Actual (scipy.solve_ivp)"),
    (euler_explicit, "Explicit Euler"),
    (euler_implicit, "Implicit Euler"),
    (midpoint, "Midpoint"),
    (trapezoidal, "Trapezoidal"),
    (adams_bashforth2, "Adams-Bashforth 2"),
    (explicit_RK2, "Explicit RK2"),
    (explicit_RK4, "Explicit RK4")
]

actual_yf = solve_ivp(dydx, [0,1], [y_init], rtol=10**(-12), atol=10**(-12)).y[0][-1]
power_law = lambda x, a, b : a*np.power(x, b)

ls="-"
index = 0
for ax_row in axes:
    for ax in ax_row:
        method, method_name = methods[index]
        ax.set_title(f"{method_name} Method", fontsize=20)
        
        if index == 0: 
            ax.text(0.27, 0.4, "(Baseline)", fontsize=40)
            index += 1
            continue # Skip the actual plot
        
        errors = np.zeros(h_steps.shape)
        for h_index, hi in enumerate(h_steps):
            xi_pts = np.arange(0, 1 + (np.float64(1)/hi), np.float64(1.0)/hi)
            yf = method(xi_pts, dydx, np.float64(1)/hi, y_init)[-1]
            errors[h_index] = abs(actual_yf - yf)
        pars, cov = curve_fit(f=power_law, xdata=h_steps, ydata=errors)
        
        #fit_c, fit_alpha = poly.polyfit(h_steps, errors, 1)
        print(f"{method_name} {pars}")
        #fit_c, fit_alpha = poly.polyfit(h_steps, errors, 1)
        ax.plot(h_steps, pars[0]*(h_steps**pars[1]), color="Red", linestyle="--", lw=1, label="Least Squares Regression Fit")
        
        ax.scatter(h_steps, errors, color="Black")
        ax.set_title(f"{method_name} Method", fontsize=20)
        ax.grid()
        ax.semilogy()
        ax.semilogx()
        ax.legend()
        ax.set_ylabel("Error", fontsize=18)
        ax.set_xlabel("Step Size (1/h)", fontsize=18)
        ax.tick_params(axis='both', which='major', labelsize=15)

        index += 1
f.tight_layout()
```

```python
def count_euler_explicit(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    count = 0
    for i in range(0, len(x) - 1):
        y[i+1] = y[i] + h*f(x[i], y[i])
        count += 1
    return y, count


def count_euler_implicit(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    count = 0
    for i in range(1, len(x) - 1):
        residual = lambda yi : yi - y[i] - h*f(x[i+1], yi)
        y_solved = fsolve(residual, y[i] + h*f(x[i], y[i]))
        y[i+1] = y[i] + h*f(x[i+1], y_solved)
        count += 3
    return y, count


def count_midpoint(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    count = 0
    for i in range(0, len(x) - 1):
        y1 = f(x[i], y[i])
        y2 = f(x[i] + h/2, y[i] + (h/2)*y1)
        y[i+1] = y[i] + h*y2
        count += 2
    return y, count


def count_trapezoidal(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    count = 0
    for i in range(0, len(x) - 1):
        residual = lambda yi : yi - y[i-1] - h*f(x[i+1], yi)
        y_solved = fsolve(residual, y[i] + h*f(x[i], y[i]))
        
        y[i+1] = y[i] + (h/2)*(f(x[i], y[i]) + f(x[i+1], y_solved))
        count += 4
    return y, count

def count_adams_bashforth2(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    y[1] = y[0] + h*f(x[0], y[0])
    count = 1
    for i in range(1, len(x) - 1):
        y[i+1] = y[i] + h*((3/2)*f(x[i], y[i]) - (1/2)*f(x[i-1], y[i-1]))
        count += 2
    return y, count


def count_explicit_RK2(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    count = 0
    for i in range(0, len(x) - 1):
        xi = x[i] + (h/2)
        yi = y[i] + (h/2)*f(x[i], y[i])
        y[i+1] = y[i] + h*f(xi, yi)
        count += 2
    return y, count


def count_explicit_RK4(x, f, h, y0):
    y = np.zeros(x.shape)
    y[0] = y0
    count = 0
    for i in range(0, len(x) - 1):
        y1 = f(x[i], y[i])
        y2 = f(x[i] + (h/2), y[i] + (h/2)*y1)
        y3 = f(x[i] + (h/2), y[i] + (h/2)*y2)
        y4 = f(x[i] + h, y[i] + h*y3)
        y[i+1] = y[i] + (h/6)*(y1 + 2*y2 + 2*y3 + y4)
        count += 4
    return y, count

methods = [
    (solve_ivp, "Actual (scipy.solve_ivp)"),
    (count_euler_explicit, "Explicit Euler"),
    (count_euler_implicit, "Implicit Euler"),
    (count_midpoint, "Midpoint"),
    (count_trapezoidal, "Trapezoidal"),
    (count_adams_bashforth2, "Adams-Bashforth 2"),
    (count_explicit_RK2, "Explicit RK2"),
    (count_explicit_RK4, "Explicit RK4")
]

f, ax = plt.subplots(1, 1, figsize=(14, 8), facecolor='w')
f.suptitle(f"Global Error and Function Calls for Each Method", fontsize=30, y=1.0)

h_steps = np.array([2**n for n in range(5, 15)])

actual_yf = solve_ivp(dydx, [0,1], [y_init], rtol=10**-9, atol=10**-9).y[0][-1]

ls="-"
colors = [
    "Blue",
    "Red",
    "Green",
    "Orange",
    "Purple",
    "Black",
    "Cyan"
]
for index, (method, method_name) in enumerate(methods[1:]):
    errors = np.zeros(h_steps.shape)
    named=False
    for hi in h_steps:
        xi_pts = np.arange(0, 1 + (np.float64(1)/hi), np.float64(1.0)/hi)
        y, M = method(xi_pts, dydx, np.float64(1)/hi, y_init)
        error = abs(actual_yf - y[-1])
        if not named:
            ax.scatter(M, error, color=colors[index], label=method_name)
            named=True
        else:
            ax.scatter(M, error, color=colors[index])

ax.grid()
ax.legend()
ax.semilogy()
ax.semilogx()
ax.set_ylabel("Error", fontsize=18)
ax.set_xlabel("Function Calls (M)", fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=15)

f.tight_layout()
```

```python
N = 10
T = np.zeros((N, N)) - 2*np.eye(N) + np.eye(N, k=1) + np.eye(N, k=-1)
eigenvalues = np.linalg.eig(T)[0]
eigenvalues.sort()
eigenvalues = eigenvalues[::-1]
f, ax1 = plt.subplots(1, 1, figsize=(14, 8), facecolor='w')
f.suptitle(f"T Matrix Eigenvalues Compared with Function", fontsize=30, y=1.0)

ax1.scatter([i for i in range(1, N+1)], eigenvalues, color="Black", label="Eigenvalues", s=70)

func = lambda x : -2*(1 - np.cos((np.pi*x) / (N+1)))
x_pts = np.arange(1, N + 0.1, 0.1)
ax1.plot(x_pts, func(x_pts), color="Blue", label="Function")
ax1.grid()
ax1.set_xlim(1, 10)
ax1.set_ylim(-4, 0)
ax1.set_ylabel("Eigenvalue", fontsize=18)
ax1.set_xlabel("i", fontsize=18)
ax1.tick_params(axis='both', which='major', labelsize=15)
ax1.legend(fontsize=15)
f.tight_layout()
```
